```python
import pandas as pd
from matplotlib  import pyplot as pyplt

data1 = pd.read_csv("movie1.csv")

pyplt.scatter(x=data1['rating'], y=data1['review_polarity'])
pyplt.show()
```

```python
data2 = pd.read_csv("movie2.csv")

pyplt.scatter(x=data2['rating'], y=data2['review_polarity'])
pyplt.show()
```

```python
data3 = pd.read_csv("movie3.csv")

pyplt.scatter(x=data3['rating'], y=data3['review_polarity'])
pyplt.show()
```

```python
data = pd.concat([data1, data2, data3], axis = 0)

pyplt.scatter(x=data['rating'], y=data['review_polarity'])
pyplt.show()
```
